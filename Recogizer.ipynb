# Folgende Code zeigt die Anzahl von Parkverletzungen in zwei verschiedene Interval

```python
zeit = df['Violation Time']
am = 0
pm = 0
for zeile in zeit:
    if type(zeile) == str:
        if 'A' in zeile:
            am += 1
        elif 'P' in zeile:
            pm += 1
print("AM:", am)
print("PM:", pm)
```
